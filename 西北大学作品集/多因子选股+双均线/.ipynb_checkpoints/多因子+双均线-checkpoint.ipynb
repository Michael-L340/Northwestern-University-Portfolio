# 选股策略

```python
hs300 = pd.read_csv('clean1.csv')

hs300['trade_date'] = pd.to_datetime(hs300['trade_date'],format='%Y%m%d')  #转化日期数据
hs300 = hs300.sort_index(ascending=False)          #时间降序排列                        
hs300.set_index('trade_date', inplace=True)         #设置index


#获取时间列表（为后续回测准备）
trade_series = hs300.index.unique().to_list()
trade_series = sorted(trade_series, reverse=False) #时间降序排列


#获取股票列表
code_list = hs300['ts_code'].unique()
```

### 计算每个股票每一天的移动平均

```python
hs300 = hs300.reset_index(drop=False)  #重置索引


#计算个股移动平均
hs300['ma5'] = hs300.groupby('ts_code')['close'].rolling(window=5, min_periods=1).mean().reset_index(level=0, drop=True)
hs300['ma30'] = hs300.groupby('ts_code')['close'].rolling(window=30, min_periods=1).mean().reset_index(level=0, drop=True)

hs300.set_index('trade_date', inplace=True)         #设置index
hs300
```

```python
#移动平均后时间匹配
#trade_series = trade_series[30:]  #删除前30天

#hs300数据进行同样操作
#hs300 = hs300[hs300.index.isin(trade_series)]

#判断ma5是否大于ma30
hs300['ma5>ma30'] = hs300['ma5'] > hs300['ma30']
```

```python
##计算个股金叉死叉时间点

hs300 = hs300.reset_index(drop=False)  #重置索引

hs300['gold'] = hs300.groupby('ts_code').apply( lambda x: (x['ma5>ma30'] & (~x['ma5>ma30']).shift(1))).reset_index(level=0, drop=True)
hs300['death'] = hs300.groupby('ts_code').apply( lambda x: (~(x['ma5>ma30'] | (~x['ma5>ma30']).shift(1)))).reset_index(level=0, drop=True)

hs300.set_index('trade_date', inplace=True)         #设置index
```

### 数据处理

```python
#流通市值自然对数
hs300['log_mv'] = np.log(hs300['circ_mv'])
hs300.columns
```

```python
factors = ['pe_ttm', 'pb','ps_ttm', 'log_mv']
#归一化处理
for factor in factors:
    hs300[factor] = (hs300[factor]- hs300[factor].min()) / (hs300[factor].max() - hs300[factor].min())
hs300
```

```python
hs300 = hs300[['ts_code', 'close', 'open','pe_ttm', 'pb','ps_ttm','log_mv','ma5', 'ma30','ma5>ma30', 'gold', 'death']]

#设置权重（权重根据一篇描述组合风险的研报）
hs300['score'] = 1*hs300['log_mv'] + 1/3*hs300['pe_ttm'] + 1/3*hs300['pb'] + 1/3*hs300['ps_ttm'] 
hs300
```

# 回测框架

```python
initial_money = 100*10000 #金额100w
cash = initial_money
number = 15 ##最大持仓


#设置仓位记录
position = pd.DataFrame(0, index=trade_series, columns=code_list)

#上一期仓位入选名单
last_stock_list = []
#记录仓位价值
portfolio = pd.DataFrame(0, index = trade_series, columns = ['value'])

for i in range(0, len(trade_series), 10):    #10天回测
    #获得得分前20的股票名字
    # i 为隐索引，获得日期方式为 trade_series[i]
    current_data = hs300.loc[trade_series[i]]       #回测日当天截面数据
    stock_list = current_data.sort_values(by = 'score', ascending =False)['ts_code'].tolist()[:number]
    
    
    
    
    #对于这期不在list的上期股票，以当天开盘价卖出
    for stock in last_stock_list:
        if stock not in stock_list:            
            
            ######################################第七次修改，[0]，避免获取series
            open_price = current_data[current_data['ts_code']==stock]['open'][0]
            ######################################
            


            
            ##########第六次修改，时间轮动
            cash += open_price* position.loc[trade_series[i -1]][stock] 
            ##############################
            position.loc[trade_series[i]][stock] = 0
            
            
    ######################################################################################
    #修改二
    #没有考虑last_stock_list
    last_stock_list = stock_list
    
    ######################################################################################
    
    ########第七次修改########
    ##range应该0-10 而不是0-11
    for j in range(0,10):                           #一个回测期周期10天
        
        ##修改一 避免i+j大于最大数
        if i+j >= len(trade_series):
            break
        
        
        #################################################################
        ###########第六次修改  设置当天初始仓位等于前一天仓位，
        position.loc[trade_series[i+j]] = position.loc[trade_series[i+j-1]]
        #################################################################
        
        
        trade_data = hs300.loc[trade_series[i+j]]  #获取回测期间内每一天的数据，根据金叉死叉的True False获取对应给股票名称，进行买卖
        trade_data = trade_data[trade_data['ts_code'].isin(stock_list)]
        
        
        gold_list = trade_data[trade_data['gold'] == True]['ts_code'].tolist()    #应当买入卖出的股票代码
        death_list = trade_data[trade_data['death'] == True]['ts_code'].tolist()
        
        ###################################################################################
        ########第三次  200改为i+j  打字失误
        
        #计算空仓股票数量
        #empty = 20 - (position.loc[trade_series[200]] != 0).sum()
        empty = number - (position.loc[trade_series[i+j]] != 0).sum()
        ###################################################################################
        
        #######第八次修改，当empty=0，应该分配金额为0
        #每股分配金额
        
        if empty ==0:
            dividend_cash=0
        if empty != 0 :
            dividend_cash = cash/empty
        
        ##################################################################################################
        #第四次修改
        ##trade_data[trade_data['ts_code'] == stock]['open']返回的是series，不是value
        
        
        
        #买入金叉
        for stock in gold_list:
            price = trade_data[trade_data['ts_code'] == stock]['open'].values[0] *100   #1手100股
            amount = dividend_cash // price
            cash -= (price * amount)
            position.loc[trade_series[i+j]][stock] = amount
            
        #卖出死叉
        for stock in death_list:            
            price = trade_data[trade_data['ts_code'] == stock]['open'].values[0] *100   #1手100股
            amount = position.loc[trade_series[i+j]][stock]
            cash += (price * amount)
            position.loc[trade_series[i+j]][stock] = 0
        
        
        #计算当天组合的价值
        portfolio_value = 0
        for stock in stock_list:
            #portfolio_value += position.loc[trade_series[i+j]][stock] * trade_data[trade_data['ts_code'] == stock]['close'].values[0]
            
                # 过滤出当前股票的数据
            stock_data = trade_data[trade_data['ts_code'] == stock]
    
            # 确保找到了对应的股票数据
            if not stock_data.empty:
                # 获取该股票的收盘价
                close_price = stock_data['close'].values[0]
        ####################################################################################################
        
#######第五次修改
###仓位价值应该乘以100

                # 计算仓位价值
                #portfolio_value += position.loc[trade_series[i+j], stock] * close_price
                portfolio_value += position.loc[trade_series[i+j], stock] * close_price*100
        
######################
            else:
                # 如果该股票数据为空，打印出来
                print(f"Warning: No data found for stock {stock} on {trade_series[i+j]}")
            
            
        portfolio_value = cash +portfolio_value 
        portfolio.loc[trade_series[i+j]]['value'] = portfolio_value
            
        ##################################################################################################    


    
portfolio['value'] = (portfolio['value']/initial_money -1)*100
portfolio
```
