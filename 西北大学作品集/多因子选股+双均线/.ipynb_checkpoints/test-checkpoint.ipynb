```python
#导入已准备好的数据
hs300 = pd.read_csv('hs300.csv')[['con_code', 'trade_date']]
hs300_stock_info = pd.read_csv('stock_info.csv')
hs300_stock_info_open = pd.read_csv('open_price.csv')

hs300 = pd.merge(hs300_stock_info, hs300_stock_info_open, how = 'outer')




data = pd.read_csv('ST.csv')

data1 = data[data['list_date'] > 20220101] #剔除上市时间小于3年
code_st = data1['ts_code'].to_list()
code_st

for i in code_st:
    hs300 = hs300[hs300['ts_code'] != i]
hs300

hs300 = hs300[['ts_code', 'close', 'open','pe_ttm', 'pb','ps_ttm','turnover_rate_f']]
```

```python
aaa = hs300[hs300.isna().any(axis=1)]['ts_code']
bbb = hs300[~hs300['ts_code'].isin(aaa)]
len(bbb['ts_code'].unique())
```
